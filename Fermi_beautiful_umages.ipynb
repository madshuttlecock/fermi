```python
import os
import random
import matplotlib.gridspec as gs
import cv2 as cv2
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
def openAsImage(fname):        
    fits_image = fits.open(fname)
    #print(fits_image.info())
    return fits_image[0].data
    
```

**Можно брать логарифм, корень, степень от картинки**

```python
def produceImage(p, ax, title, cmap_name = 'CMRmap', minus = False, power = False, log = False, sqrt=False):
    name, scale = p
    mm = openAsImage(name)
    mm = mm.astype(np.float32)
    #print(mm.min(), mm.mean(), mm.max())

    if power:
        mm = mm**scale
    if log:
        mm = np.log(1 + mm)
        mm = np.log(1 + mm)
    if sqrt:
        mm = np.sqrt(mm)
    if minus:
        mm = -mm
        
    #mm = np.log(mm)
    ax.imshow(mm, cmap=cmap_name)
    ax.set_title(title, fontsize=18)
    ax.tick_params(labelbottom=False, labelleft=False)
```

```python
def drawImages(name, cmap = 'CMRmap', minus=False, power=False, log=False, sqrt=False):
    name_mm = name
    name_dm = name[:-6] + "2.fits"
    titles = ['model_map', 'deconvolved_map']
    images = [(name_mm, 2), (name_dm, 1)]

    i = 0
    #print(images[i])
    fig, axs = plt.subplots(figsize=(20,20),nrows=1, ncols=2, constrained_layout=False)
    for ax in axs.flatten():
        produceImage(images[i], ax, titles[i], cmap, minus, power, log, sqrt)
        i = i + 1
    plt.suptitle(name[name.find('_')+1:-15], fontsize = 30)        
    plt.show()
```
